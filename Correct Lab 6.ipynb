```python
import pandas as pd
from pandas import *
import numpy as np

df = pd.read_csv("mydata.csv")

df.head(6)

## Problem 1
median = df['median_house_value'].median()
print(f"Median house value {median}")

## Problem 2
median1 = df['total_rooms'].median()
print(f"Median total rooms value {median1}")

for x in df.index:
  if df.loc[x, "total_rooms"] < 500:
    df.loc[x, "total_rooms"] = median1

## Problem 3
for x in df.index:
 if df.loc[x, "housing_median_age"] < 25:
    print(df.loc[x])

## Problem 4
import matplotlib.pyplot as plt

df.plot(kind = 'scatter', x = 'total_bedrooms', y = 'total_rooms')

x = df['total_bedrooms']
y = df['total_rooms']

# Fit a linear regression line
coef = np.polyfit(x, y, 1)
poly_fn = np.poly1d(coef)

# Create smooth x-range for the line
xs = np.linspace(x.min(), x.max(), 100)

# Plot trendline
plt.plot(xs, poly_fn(xs), color='red')
plt.show()
```
